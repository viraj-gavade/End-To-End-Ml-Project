## Data Cleaning

```python
#missing values 
dataset[dataset.isnull().any(axis=1)]
```

The datas set is didvided into two regions

```python
dataset.loc[:122,'Region'] = 0
dataset.loc[122:,'Region'] = 1
```

```python
dataset.info()
```

```python
dataset['Region'] = dataset['Region'].astype(int)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
# Removing the null and nan values
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
## Fix the spaces in column names
dataset.columns=dataset.columns.str.strip()
```

```python
dataset.columns
```

```python
dataset.info()
```

```python
dataset.head()
```

## Changing the data type of coulmns to integer

```python
## Change the required column as integer data type
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]] = dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]].astype(int)
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
## Changing the othet columns to the data type
objects = [features for features in dataset.columns if dataset[features].dtypes == 'O']
```

```python
for i in objects:
    if(i!='Classes'):
        dataset[i] = dataset[i].astype(float)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
# Saving the cleaned data set 

dataset.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```
